```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# standard temperature 0°C
θ_0 = 273.15 # Kelvin

# temperature range: 0°C to 200°C 
θ = np.linspace(θ_0, θ_0+200, num=500)

# Carnot efficiency
η = (θ - θ_0) / θ
```

```python
fig, ax = plt.subplots(dpi=200)
ax.plot(θ, η);
```
